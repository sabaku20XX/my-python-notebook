```python
import pandas
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
dataset = pandas.DataFrame(boston.data, columns=boston.feature_names)
dataset['target'] = boston.target
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
dataset.iloc[:, :] = scaler.fit_transform(dataset.values)

dataset.head()
```
